## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the Dataset

```python
dataset_df = pd.read_csv('3.K_Means_Clustering.csv')
```

```python
dataset_df
```

```python
x = dataset_df.iloc[:,[3,4]].values
```

```python
x
```

## K Means Clustering

```python
import os
os.environ["OMP_NUM_THREADS"] = "1"
```

```python
from sklearn.cluster import KMeans
```





## WCSS -> within cluster sum squares



## Use the elbow method to find the optimal number of clusters

```python
wcss = []
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters = i , init = 'k-means++' , random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)   
```

```python
plt.plot([i for i in range(1,11)],wcss,color='purple')
plt.title('Elbow Method for Kmeans')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid()
plt.plot()
```

## Training the KMeans model on the dataset

```python
kmeans = KMeans(n_clusters=5,init='k-means++',max_iter=300,random_state=42)
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

## Visualizing the clusters

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='green',label='Cluster 3')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=100,c='cyan',label='Cluster 4')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=100,c='magenta',label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Cluster Centroid')
plt.title('K Means clustering of Annual Earning & Spending Score')
plt.xlabel('Annual Earnings (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid()
plt.show()
```
