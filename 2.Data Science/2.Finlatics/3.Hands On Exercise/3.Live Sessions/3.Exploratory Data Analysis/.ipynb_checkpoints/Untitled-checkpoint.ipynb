```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Removing Unwanted Columns

```python
dataset_df.drop(columns=['Unnamed: 0','Track ID','Album ID'],inplace=True)
```

```python
dataset_df.info()
```

```python
index_to_drop = dataset_df[dataset_df['Popularity'] == 0].index
dataset_df.drop(index_to_drop,inplace=True)
```

```python
dataset_df['Duration (min)'] = dataset_df['Duration (ms)']/(1000*60)
dataset_df['Explicit'] = dataset_df['Explicit'].astype(int)
```

```python
dataset_df['Release Date'] = pd.to_datetime(dataset_df['Release Date'],errors='coerce')
dataset_df['Release Year'] = dataset_df['Release Date'].dt.year
```

```python
dataset_df.info()
```

```python
numeric_df = dataset_df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(12,8))
sns.heatmap(data=correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

## How does the release date (year) affect track popularity?

```python
avg_popularity_by_year = dataset_df.groupby('Release Year')['Popularity'].mean().reset_index()
plt.figure(figsize=(6,4))
sns.lineplot(avg_popularity_by_year,x='Release Year',y='Popularity',color='purple')
plt.title('Track Release Year vs Popularity')
plt.xlabel('Release Year')
plt.ylabel('Popularity')
plt.show()
```

```python
avg_popularity_by_year = dataset_df.groupby('Release Year')['Popularity'].mean().reset_index()
window_size = 5
avg_popularity_by_year['Smoothed_Popularity_MA'] = avg_popularity_by_year['Popularity'].rolling(window=window_size,center=True,min_periods=1).mean()
plt.figure(figsize=(6,4))
sns.lineplot(avg_popularity_by_year,x='Release Year',y='Popularity',color='purple')
sns.lineplot(avg_popularity_by_year,x='Release Year',y='Smoothed_Popularity_MA',color='red')
plt.title('Track Release Year vs Popularity')
plt.xlabel('Release Year')
plt.ylabel('Popularity')
plt.grid()
plt.show()
```

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
avg_popularity_by_year = dataset_df.groupby('Release Year')['Popularity'].mean().reset_index()

# LOWESS smoothing
lowess_fraction = 0.1
smoothed_lowess = lowess(
    avg_popularity_by_year['Popularity'],
    avg_popularity_by_year['Release Year'],
    frac=lowess_fraction
)
avg_popularity_by_year['Smoothed Popularity Lowess'] = smoothed_lowess[:, 1]

# Plot
plt.figure(figsize=(6, 4))
sns.lineplot(data=avg_popularity_by_year, x='Release Year', y='Popularity', color='purple', label='Original')
sns.lineplot(data=avg_popularity_by_year, x='Release Year', y='Smoothed Popularity Lowess', color='red', label='LOWESS Smoothed')

plt.title('Track Release Year vs Popularity')
plt.xlabel('Release Year')
plt.ylabel('Popularity')
plt.grid()
plt.legend()
plt.show()
```

## What are the most common tempo ranges, and do specific tempos align with more popular tracks?

```python
plt.figure(figsize=(6,4))
sns.histplot(data=dataset_df['Tempo'],bins=30,kde=True,color='purple')
plt.title('Histogram Distribution of Tempo')
plt.xlabel('Tempo')
plt.ylabel('Frequency')
plt.show()
```

```python
bins = [0,80,110,140,170,np.inf]
labels = ['<80 (Slow)','80-110 (Moderate)','110-140 (Upbeat)','140-170 {Fast}','>170 {Very Fast}']
dataset_df['Tempo Category'] = pd.cut(dataset_df['Tempo'],bins=bins,labels=labels,right=False)
dataset_df['Tempo Category'].value_counts().sort_index()
```

```python
avg_popularity_by_tempo = dataset_df.groupby(dataset_df['Tempo Category'])['Popularity'].mean().reset_index()
```

```python
sns.barplot(avg_popularity_by_tempo,x='Tempo Category',y='Popularity',palette='viridis',hue='Tempo Category')
plt.title('Average Popularity by Tempo Category')
plt.xlabel('Tempo Category (BPM)')
plt.ylabel('Popularity')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(dataset_df,x='Tempo Category',y='Popularity',palette='viridis',hue='Tempo Category')
plt.title('Average Popularity by Tempo Category')
plt.xlabel('Tempo Category (BPM)')
plt.ylabel('Popularity')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

## Are certain artists or albums consistently producing popular tracks?

```python
dataset_df_artists = dataset_df.assign(Artists=dataset_df['Artists'].str.split(', ')).explode('Artists')
dataset_df_artists['Artists'] = dataset_df_artists['Artists'].str.strip()
```

```python
dataset_df_artists['Artists']
```

```python
artists_track_counts = dataset_df_artists.groupby('Artists').size()
```

```python
artists_track_counts
```

```python
artists_with_6_or_more_tracks = artists_track_counts[artists_track_counts >= 6].index.tolist()
```

```python
dataset_df_artists_filtered = dataset_df_artists[dataset_df_artists['Artists'].isin(artists_with_6_or_more_tracks)]
artist_pop_filtered = dataset_df_artists_filtered.groupby('Artists')['Popularity'].mean().reset_index()
top_artists = artist_pop_filtered.sort_values(by='Popularity',ascending=False).head(10)
```

```python
top_artists
```

```python
plt.figure(figsize=(15,4))
sns.barplot(data=top_artists,x='Artists',y='Popularity',errorbar=None)
plt.title('Average Popularity by Artists Flag (Tracks >= 6)')
plt.xlabel('Artists')
plt.ylabel('Average Popularity')
plt.show()
```
