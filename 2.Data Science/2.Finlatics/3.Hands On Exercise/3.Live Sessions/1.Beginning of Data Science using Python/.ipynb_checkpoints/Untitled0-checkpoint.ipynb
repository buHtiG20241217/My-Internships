## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Importing the Dataset

```python
dataset_df = pd.read_csv('music_data.csv')
```

```python
dataset_df.head()
```

```python
dataset_df.info()
```

```python
dataset_df.isna().sum()
```

```python
dataset_df.info()
```

## Getting a detailed summary

```python
mean_values = dataset_df.mean(numeric_only=True)
median_values = dataset_df.median(numeric_only=True)
mode_values = dataset_df.mode(numeric_only=True).iloc[0]
variance_values = dataset_df.var(numeric_only=True)
std_dev = dataset_df.std(numeric_only=True)
numeric_columns = dataset_df.select_dtypes(include=['int64','float64']).columns
range_values = dataset_df[numeric_columns].max() - dataset_df[numeric_columns].min()
percentiles = dataset_df[numeric_columns].quantile([0.25,0.5,0.75])
IQR = percentiles.loc[0.75] - percentiles.loc[0.25]

print(f"Numeric Columns\n{numeric_columns}\n")
print('\nStatistical summary of numeric columns\n')
print(f"1.Mean Values\n{mean_values}\n")
print(f"2.Median Values\n{median_values}\n")
print(f"3.Mode Values\n{mode_values}\n")
print(f"4.Variance Values\n{variance_values}\n")
print(f"5.Standard Deviation Values\n{std_dev}\n")
print(f"6.Range Values\n{range_values}\n")
print(f"7.Percentiles\n{percentiles}\n")
print(f"8.Inter Quantile Ranges(IQR)\n{IQR}\n")
```

## Checking for skewness

```python
3*(mean_values-median_values) == (mean_values-mode_values)
```

## Creating a new column

```python
dataset_df['Release Date'] = pd.to_datetime(dataset_df['Release Date'],errors='coerce')
dataset_df['Release Year'] = dataset_df['Release Date'].dt.year
```

## Dropping unnecessary columns

```python
dataset_df = dataset_df.drop(columns=['Unnamed: 0','Track Name','Album Name','Album ID','Track ID'])
```

```python
dataset_df.info()
```

```python
dataset_df['Duration (min)'] = dataset_df['Duration (ms)'] / (1000*60)
```

## Visualizing Dataset

```python
plt.figure(figsize=(10,6))
plt.hist(dataset_df['Popularity'],bins=30,color='purple',edgecolor='black')
plt.title('Distribution of popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

## Deleting missing data

```python
index_to_drop = dataset_df[dataset_df['Popularity'] == 0].index
dataset_df.drop(index_to_drop,inplace=True)
```

```python
dataset_df.info()
```

## Recalculating summary

```python
mean_values = dataset_df.mean(numeric_only=True)
median_values = dataset_df.median(numeric_only=True)
mode_values = dataset_df.mode(numeric_only=True).iloc[0]
variance_values = dataset_df.var(numeric_only=True)
std_dev = dataset_df.std(numeric_only=True)
numeric_columns = dataset_df.select_dtypes(include=['int64','float64']).columns
range_values = dataset_df[numeric_columns].max() - dataset_df[numeric_columns].min()
percentiles = dataset_df[numeric_columns].quantile([0.25,0.5,0.75])
IQR = percentiles.loc[0.75] - percentiles.loc[0.25]

print(f"Numeric Columns\n{numeric_columns}\n")
print('\nStatistical summary of numeric columns\n')
print(f"1.Mean Values\n{mean_values}\n")
print(f"2.Median Values\n{median_values}\n")
print(f"3.Mode Values\n{mode_values}\n")
print(f"4.Variance Values\n{variance_values}\n")
print(f"5.Standard Deviation Values\n{std_dev}\n")
print(f"6.Range Values\n{range_values}\n")
print(f"7.Percentiles\n{percentiles}\n")
print(f"8.Inter Quantile Ranges(IQR)\n{IQR}\n")
```

## Viewing Outliers

```python
q1 = dataset_df['Popularity'].quantile(0.25)
q3 = dataset_df['Popularity'].quantile(0.75)
IQR = q3-q1
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR

outliers = dataset_df[(dataset_df['Popularity'] < lower_bound) | (dataset_df['Popularity'] > upper_bound)]
outliers.shape[0]
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=dataset_df['Popularity'],color='purple')
plt.title('Box Plot for Popularity')
plt.xlabel('Popularity')
plt.show()
```

```python
q1 = dataset_df['Duration (ms)'].quantile(0.25)
q3 = dataset_df['Duration (ms)'].quantile(0.75)
IQR = q3-q1
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR

outliers = dataset_df[(dataset_df['Duration (ms)'] < lower_bound) | (dataset_df['Duration (ms)'] > upper_bound)]
outliers.shape[0]
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=dataset_df['Duration (ms)'],color='purple')
plt.title('Box Plot for Duration (ms)')
plt.xlabel('Duration (ms)')
plt.show()
```
