```python
def edit_with_median(sex,sex_type,survived,sex_class,sex_embark_town):
    mask = (
        (dataset_df['sex'] == sex) &
        (dataset_df[sex_type]) &
        (dataset_df['survived'] == survived) &
        (dataset_df['class'] == sex_class) &
        (dataset_df['embark_town'] == sex_embark_town)
    )
    median_age = dataset_df.loc[mask, 'age'].median()
    dataset_df.loc[mask & dataset_df['age'].isna(), 'age'] = median_age
```
